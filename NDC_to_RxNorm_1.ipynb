```python
import pandas as pd
```

```python
concept=pd.read_csv("CONCEPT.csv",sep="\\t",error_bad_lines=False, converters={"concept_id":str})
```

```python
# List all vacabularies in concepts
concept["vocabulary_id"].unique()
```

```python
# Select only the needed vocabularies
concept=concept[(concept["vocabulary_id"]=='NDC')|(concept["vocabulary_id"]=='RxNorm')]
```

```python
#1 step NDC TO CONCEPT_ID(NDC)

ndc=pd.read_csv('NDC codes.csv',converters={"ndc":str})

concept_ndc=ndc.merge(concept, how='left',left_on="ndc", right_on="concept_code")

concept_ndc
```

```python
#2 step CONCEPT_ID(NDC)

concept_dic=pd.read_csv('CONCEPT_RELATIONSHIP.csv',sep="\\t",error_bad_lines=False)
```

```python
concept_dic



```

```python
# Select only relationships on mapping to. 

concept_dic=concept_dic[concept_dic['relationship_id']=="Maps to"]
concept_dic.concept_id_1=concept_dic.concept_id_1.astype(str)
concept_dic.concept_id_2=concept_dic.concept_id_2.astype(str)

concept_dic
```

```python
# 3 step. Merge tables joining on concept_id(rx_norm) to concept code rx_norm( rx_norm official code)

dic_join=concept_ndc.merge(concept_dic, how='inner',left_on="concept_id",right_on="concept_id_1")
dic_join=dic_join[['drug','ndc','concept_id_1','concept_id_2']]
dic_join
```

```python
#renaming for clarity


final_join=dic_join.merge(concept, how='inner',left_on="concept_id_2",right_on="concept_id")
final_join=final_join[['drug','ndc','concept_id_1','concept_id_2','concept_code']]
final_join

final_join.columns=['drug','ndc','concept_id_ndc','concept_id_rx_norm','rx_norm']
final_join
```

```python
final_join.to_csv("NDC_to_RxNorm.csv")
```
